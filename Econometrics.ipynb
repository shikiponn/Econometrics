### 小標本

OLSの5つの仮定

仮定1(条件付き期待値Ver.).  誤差項と変数は独立$y_i - x'_i\beta + u_i$  
仮定1.  無相関  
仮定2（標本Ver.）.  観測データはフルランク  
仮定2.  期待値はフルランク  
仮定3.  条件付き均一分散  
仮定3. 分散の期待値は一定  
仮定4. 　 観測値はi.i.d.  

#### 推定量
  
$$\hat{\beta} =  (X'X)^{-1}X'(X\beta + U) = \beta + (X'X)^{-1}X'U$$  

##### 残差回帰（Auxiliary regression）
もしくは残差回帰を利用して
$$\hat{\beta}_j = \frac{\Sigma \hat{v}_{ji} y_i } {\Sigma \hat{v}_{ji}^2}  
= \beta_j + \frac{\Sigma \hat{v}_{ji} u_i } {\Sigma \hat{v}_{ji}^2}
$$

#### 推定量の不偏性、条件付き分散
・推定量の条件付き期待値：$E(\beta | X) = \hat{\beta} - \beta = \beta $ （$\because$仮定より$E(U | X) = E(u_i | x_i) = 0$）
※要チェック  
不偏性には条件付き均一分散の仮定は必要ない

・条件付き分散（均一分散）
$$Var(\hat{\beta}|X) = E[((\beta + (X'X)^{-1}X'U) - E[\beta + (X'X)^{-1}X'U])^2 | X] \\
= E[((X'X)^{-1}X'U)^2] \\
= E[(X'X)^{-1}X'UU'X(X'X)^{-1} | X] \\
= (X'X)^{-1}X'E(UU'|X)X(X'X)^{-1} \\
= \sigma^2(X'X)^{-1}X'X(X'X)^{-1} \\
= \sigma^2(X'X)^{-1}
$$

### 仮説検定

第一種の過誤：帰無仮説を採択すべき時に棄却する確率 = 有意水準   
第二種の過誤：帰無仮説を棄却すべき時に棄却しない確率  

仮説検定には
+ Wald検定
+ 尤度比検定
+ ラグランジュ乗数検定
の3つがある。線形回帰モデルの係数に対する帰無仮説が適切に推定された$\sigma^2$が使われれば、この3つは一致する。

有意水準が示すものは帰無仮説が正しい時にこれを棄却する回数の割合を示す。

#### t検定
t検定統計量は
$$ \frac{N(0, 1)}{\sqrt{\chi^2(df)/df}} \sim t(df)$$
但し分母と分子は独立（cf. Hayashi p.37）  
二次元正規分布が無相関なら独立という特殊なケースを用いてる


$$
t =\frac{\hat{\beta}_j - \beta_j}{SE(\hat{\beta}_j)} \\
= \frac{\hat{\beta}_j - \beta_j}{\sqrt{  \hat{sigma}^2(X'X)^{-1}_{jj}  }} \sim t(N - K)  \\
=\frac{\hat{\beta}_j - \beta_j}{\sqrt{  \sigma^2(X'X)^{-1}_{jj}  }} \times \frac{\sqrt{\sigma^2}}{\sqrt{\hat{\sigma}^2}}  \\
=\frac{z}{\sqrt{  \hat{\sigma}^2 / \sigma^2 }}  \\
= z \times \frac{1}{ \sqrt{      \frac{\chi^2(n-K)}{n-K}    }       }
$$
通常母分散は不明なので標本（不偏）分散で代用する。

t分布の中心が0でない時、非心t分布（Non-centred t- distribution）

#### 定式化の失敗
##### 過剰定式化
母集団において係数がゼロの変数を入れてしまうこと。追加された変数が他の変数と相関しているかは問題ではない。
つまり$x_3$が過剰な変数だとしたら  $$E(y | x_1, x_2, x_3) = E(y | x_1, x_2) = \beta_0 + \beta_1x_1 + \beta_2x_2$$  
となることである。
しかし$x_3$残差と相関している場合は他の係数にバイアスを与える。

- 係数は不偏
- 誤差は増大
（精度減少）
- 推定結果を用いた統計的推論は妥当
これは追加された変数の係数をゼロと推定していると考えれば納得がいく。

##### 過小定式化
必要な変数を入れずに回帰してしまうこと。変数がそもそもデータに無かったり高次の多項式を省略した時に起きる。
$$E(y | x, z) = x'\beta_0 + z'\gamma_0 $$  
で$z'$を入れ忘れた時、残差は$z'\gamma_0  + \epsilon$になる。この時zとx(入れ忘れた変数と残りの全ての変数)が無相関なら残差$z'\gamma_0  + \epsilon$とｚも無相関で推定量$\beta_0$は一致性を満たす。$\because E(\epsilon | x)= 0 $from the assumption

過小定式化で入れ忘れた変数が他の変数と相関を持つならば・・・
* 係数は不偏、一致ではなくなる
* 偏りの方向は純粋な効果と相関に依存、符号が変わることも。補助回帰によって調べることができる。
* 統計的推論は誤ったものに

#### 必要な変数を多めに入れた場合
分散が小さくなるので影響のある変数はなるべく入れたほうが良い。

#### 観測（測定）誤差
##### 従属変数（被説明変数）の誤差（Errors in the dependent variables）
この場合必ずしも $E(\epsilon | x)= 0 $ が満たされないとは限らない。
$$
y* = x'\beta_0 + \epsilon  \\
y= y* + v
$$
のとき
$$ y = x'\beta_0 + \epsilon + v $$
もし $E(v | x)= 0 $ なら観測誤差はないものとして扱える。しかし $E(v | x) \neq 0 $なら$\beta_0$は一致性を満たさない。 

##### 独立変数（説明変数）の誤差（Errors in the regressors）
この場合正しい識別条件は満たされない。
$$
y = x'\beta_0 + z*'\gamma_0 +  \epsilon  \\
z = z* + v  \text{  where $E(z | x*, z*)= z*$}
$$
のとき、観測されたy, x, zをつかって以下の式を推定するに等しい。
$$
y = x'\beta_0 + z'\gamma_0 +  \epsilon - v'\gamma_0
$$
vとx、εとxは無相関なのでxと$\epsilon - v'\gamma_0$も無相関である。
加えて$z*$と$\epsilon$は無相関なので$\epsilon$が無相関ならzと$\epsilon$も無相関である
しかし z と $- v'\gamma_0$ は相関している。なぜならvと $v'\gamma_0$ は $v'\gamma_0 = 0$ かVar(v) = 0出ないかぎり相関しているからである。

説明変数に観測誤差がある場合ゼロ方向に不一致性がある。これをAttenuation効果と呼び「観測誤差により回帰係数が薄められる」などと呼ぶ。  
Attenuationは常に起こるわけではない。先に述べた様に説明変数と攪乱項に相関がない、つまり直行条件が満たされていれば回帰係数は一致性を満たす。

#### 同時方程式？同時性？（Simutaniety）

例えば需要と供給の同時方程式モデルがそれである。

#### 不偏性（Unbiasedness）

$$
E[\hat{\beta}|X] =  E[\beta | X] + (X'X)^{-1}X'E[U | X] \\
    = \beta \text{ \because  E[U | X] = 0 }
$$

#### 一致性（Consistency）
$$
\hat{\beta} = \beta + (X'X)^{-1}X'U \\
                = \beta + (N^{-1}X'X)^{-1}N^{-1}X'U \\
               \text{両辺にN/Nをかけた}
$$
そして
$$
N^{-1}X'X = N^{-1}\Sigma x_ix'_i  →p E(x_ix'_i) \\
N^{-1}X'U = N^{-1}\Sigma x_iu_i   →p E(x_iu_i) = 0
$$
よってContinuous mapping theoremより第二項はゼロに収束する
#### 漸近正規生（Asymptotic Normality）

$$
\sqrt{N}(\hat{\beta} - \beta) = \sqrt{N} (X'X)^{-1}X'U \\
                = (N^{-1}X'X)^{-1}N^{-1/2}X'U \\
$$

中心極限定理より  
$N^{-1/2}X'U = N^{-1/2}\Sigma(x_iu_i)  
= \frac{\sqrt{N}}{\sqrt{N}}N^{-1/2}\Sigma(x_iu_i) = 
\sqrt{N}[\frac{1}{N}\Sigma(x_iu_i) - E[x_iu_i]]$ →d　$N(0, Var(x_iu_i))$  
$ \because E[x_iu_i] = 0$ from the orthogonal cond.  &  
  $Var[\frac{1}{N}\Sigma(x_iu_i) - E[x_iu_i]] = 
E[(\frac{1}{N}\Sigma(x_iu_i) - E[x_iu_i])(\frac{1}{N}\Sigma(x_iu_i) - E[x_iu_i])']$  
$\therefore$ continuous mapping theorem と Slutsuky's theoremより  

$$ \sqrt{N}(\hat{\beta} - \beta) →d  N(0, E(x_ix'_i)^{-1}Var(x_iu_i)E(x_ix'_i)^{-1})$$

## 分位点回帰

今までは条件付き期待値について調べてきたがこれは条件付き分布関数の一側面にすぎない。分位点回帰を使えば条件付き分布関数を調べることができる。
解析解は得られない（たぶん繰り返し最小二乗法で解く？

推定量は一致性を持ち、漸近正規生を不均一分散化でも満たす。
誤識別に関して条件付き期待値モデルとほぼ同じことが言えるが、非説明変数の測定誤差に関しては条件がより厳しくなる。分位点回帰では条件付き分位点は動いてはならないからである。  
分位点回帰の推定量はこの場合一致性を満たさない、なぜなら条件付き残差の分布は元の残差と測定誤差の混合分布となるからだ。そしてこれは元のの分布とは必ず異なる。この２つの分布が一致するのは非常に特殊な条件下のみである。


## 見かけ上無相関な回帰方程式システム（Seemingly Unrelated Regression (Equation) (SUR/SURE) Model）

復数方程式からなるシステムに対してGMMを適用すれば個別の式に対してGMMを適用するより効率的な推定量が得られる場合がある。しかし
1. 方程式システムが丁度識別の場合
2. 全ての誤差項にない場合　$E(\epsilon_{im}\epsilon_{im'}x_{im}x_{im'}) = 0 , \forall m \neq m'$

1.の場合２つの推定量は同一のものである  
2.の場合２つの推定量は漸近的に等しい  
もしSURモデルの中の全ての説明変数が共通であれば各式のパラメータは丁度識別であるので、誤差項に相関があっても効率性は改善されない。  
また明らかに、もしモデルの中の式の一つでも正しくなければ全ての推定係数は一般に影響を受ける。
