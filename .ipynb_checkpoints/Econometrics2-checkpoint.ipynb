## Panel data
### Lec.1
#### Why panel data?
1. クロスセクション分析より柔軟なモデルを扱える
2. 分散共分散行列を推定することで効率性をあげられる
3. 有効な操作変数がなくても、固定効果モデルを用いることで時間によって異なる変数の係数が一致性を持つ  

加えてなにより
+ 動学を分析することができる  

事が大きい。


+ 以下全ての個票データがT期間分あるモデル（balanced panel data：調和パネル）を仮定して分析を行うが、現実にはそうでないもの（unbalanced panel data：非調和パネル）が多い。
+ iが個人を表し、Tが時間を表すとは限らない。　E.g. i が企業でTが産業、iが家族でTが家族の中の一人  。  
典型的にはTはNよりはるかに小さいので漸近的にはTを固定しN→∞を考える。これはパネルデータの特徴である。  
しかし最近ではNよりは遅いがTもまたT→∞という分析もある。

各期において誤差項との相関無し、均一分散を満たすとすると

全てのt期を通じて係数が不変ならまずOLSをして、次に得られた分散共分散行列を用いてGLSで求められる。

ここまで見たように誤差項が無相関ならパネルデータを使うことで効率性を上げられる。例えば分散共分散行列の中に関数間の情報が含まれるからである。
例えば
1. i が農場なら各農場ごとの土地の特性の情報が
2. 生産関数なら企業ごとの経営能力が
3. 所得関数なら個人の能力（教育と相関するかもしれない）が  
これに当たる。
1.のケースでは肥料の過家具がIVをとして使えるかもしれない。

しかし誤差項に特殊な構造を仮定しパネルデータを利用できるなら、固定効果メソッドを使える。
+ 固定効果  
個人効果が説明変数と相関していれば個人効果を「固定効果（Fixed Effect）」として扱う。

+ ランダム効果
個人効果が説明変数と無相関なら個人効果はランダムであると言い、ランダム効果モデル（random effect model）と呼ばれる。



固定効果モデルでは説明変数と相関するが時間を通じて一定の個人効果を導入する。
しかし個人効果が係数の影響を吸収するので、個人効果と時間を通じて一定の変数を分離して識別できない。
この時パラメータは回帰係数と”個人効果と時間を通じて一定の変数＊その係数”に鳴る　　
これはN→∞でも回帰係数は一致性を満たす。
